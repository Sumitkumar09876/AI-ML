```python
import pandas as pd
```

```python
dataset=pd.read_csv(r"/content/Loan payments data.csv")
dataset.head()
```

```python
dataset['Gender'].fillna(dataset['Gender'].mode()[0],inplace=True)
```

```python
dataset.isnull().sum()
```

```python
en_dataset=dataset[['Gender']]
en_dataset
```

get_dummies is a pandas function to encoding in boolean form

```python
pd.get_dummies(en_dataset)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

Transform into 0 and 1

```python
ohe=OneHotEncoder(drop='first')
ohe.fit_transform(en_dataset)
```

```python
ar=ohe.fit_transform(en_dataset).toarray()
pd.DataFrame(ar,columns=["Gender_male"])
```
